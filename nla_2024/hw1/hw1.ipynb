## Задача 1 (13 pts)

- (1 pts) Докажите, что $\| A \|_F \le \sqrt{\mathrm{rank}(A)} \| A \|_2$.
- (1 pts) Покажите, что для любых $m, n$ и $k \le \min(m, n)$ существует $A \in \mathbb{R}^{m \times n}: \mathrm{rank}(A) = k$, такая что $\| A \|_F = \sqrt{\mathrm{rank}(A)} \| A \|_2$.
- (1 pts) Докажите, что $\| A B \|_F \le \| A \|_2 \| B \|_F$.
- (1 pts) Докажите, что $\|A\|_2 \leq \sqrt{\|A\|_1 \|A\|_{\infty}}$
- (1 pts) Докажите субмультипликативность Фробениусовой нормы
- (2 pts) Пусть матрица $A \in \mathbb{C}^{n \times n}$ и её сингулярное разложение $A = U\Sigma V^*$. Найдите сингулярное разложение матрицы $\begin{bmatrix} 0 & A^* \\ A & 0 \end{bmatrix}$ размера $2n \times 2n$
- (2 pts) Докажите, что $\|UAV\|_2 = \| A\|_2$ и $\|UAV\|_F = \| A\|_F$ для квадратных унитарных матриц $U$ и $V$, при этом матрица $A$ не обязательно квадратная.
- (2 pts) Докажите, что если матрица треугольная и ортогональная, то она диагональная.
- (2 pts) Пусть $S$ квадратная действительная матрица и $S^\top = -S$. Покажите, что матрица $I - S$ обратима и матрица $(I-S)^{-1}(I + S)$ ортогональна.

## Решения
### Задача 1
$$
\| A \|_F = \sqrt{\sum_{i=1}^r{\sigma_i}} \leq \sqrt{\sum_{i=1}^r{\sigma_1}} = \sqrt{\sigma_1 r} = \sqrt{\mathrm{rank}(A)} \| A \|_2
$$
### Задача 2
Неравенство
$$
\sqrt{\sum_{i=1}^r{\sigma_i}} \leq \sqrt{\sum_{i=1}^r{\sigma_1}}
$$
переходит в равенство только в случае, когда
$$
\sigma_1 = \sigma_2 = ... = \sigma_r
$$
достигается это при
$A = U\Sigma_r V^*$, где $U, V$ - унитарные, а $\Sigma_r = \sigma_1 I$  
### Задача 3
$$
\|AB\|_F = \sqrt{\mathrm{tr}(B^\top A^\top AB)} \leq \|A\|_2 \sqrt{\mathrm{tr}(B^\top B)} = \|A\|_2\|B\|_F
$$
### Задача 4
Заметим, что
$$
\|A\|_2 = \sigma_{max}(A) = \sqrt{\left| \lambda_{max}(AA^*) \right|}
$$
Утверждение $\lambda_{max}(A) \leq \|A\|$ справедливо для любой нормы. Пусть $x$ - собственный вектор $A$, $\lambda$ - соответствующее собственное значение тогда
$$
\lambda x = Ax \\ 
\left| \lambda \right| \|x\| = \|Ax\| = \frac{\|Ax\|}{\|x\|} \|x\| \leq \sup_{x \neq 0} \left( \frac{\|Ax\|}{\|x\|} \right) \|x\| = \|A\|\|x\| \\
\left| \lambda \right| \leq \|A\|
$$
Тогда 
$$
\|A\|_2^2 = \|AA^*\| = \left| \lambda_{max} (AA^*) \right| \leq \|AA^*\|_1 \leq \|A^*\|_1 \|A\|_1 = \|A\|_{\infty} \|A\|_1 \\
\|A\|_2 \leq \sqrt{\|A\|_{\infty} \|A\|_1}
$$
### Задача 5
Необходимо доказать $\|AB\|_F \leq \|A\|_F\|B\|_F$
что равносильно $\|AB\|_F^2 \leq \|A\|_F^2\|B\|_F^2$
$$
\|AB\|_F^2 = \sum_{i=1}^n \sum_{j=1}^m (\sum_{k=1}^l{a_{ik} b_{kj}})^2 \leq \sum_{i=1}^n \sum_{j=1}^m (\sum_{k=1}^l a_{ik})^2 (\sum_{k=1}^l b_{kj})^2 = \|A\|_F^2 \|B\|_F^2
$$
где неравенство выполняется по неравенству Коши-Буняковского
### Задача 6
Пусть
$$
B = \begin{bmatrix} 0 & A^* \\ A & 0 \end{bmatrix} = \begin{bmatrix} U & 0 \\ 0 & V \end{bmatrix} \begin{bmatrix} 0 & \Sigma \\ \Sigma & 0 \end{bmatrix} \begin{bmatrix} U^* & 0 \\ 0 & V^* \end{bmatrix}
$$
Тогда
$$
BB^* = \begin{bmatrix} U & 0 \\ 0 & V \end{bmatrix} \begin{bmatrix} \Sigma^2 & 0 \\ 0 & \Sigma^2 \end{bmatrix} \begin{bmatrix} U^* & 0 \\ 0 & V^* \end{bmatrix}
$$
$$
B^*B = \begin{bmatrix} U & 0 \\ 0 & V \end{bmatrix} \begin{bmatrix} \Sigma^2 & 0 \\ 0 & \Sigma^2 \end{bmatrix} \begin{bmatrix} U^* & 0 \\ 0 & V^* \end{bmatrix}
$$
Таким образом, SVD для матрицы $B$ будет
$$
\begin{bmatrix} U & 0 \\ 0 & V \end{bmatrix} \begin{bmatrix} \Sigma & 0 \\ 0 & \Sigma \end{bmatrix} \begin{bmatrix} U^* & 0 \\ 0 & V^* \end{bmatrix}
$$
### Задача 7
По определению матричной нормы
$$
\|UAV\|_2 = \sup_{x \neq 0} \frac{\|UAVx\|_2}{\|x\|_2}
$$
Так как матрицы $V$ и $U$ унитарны сделаем замену $Vx = y$, причем их нормы равны, справедливо следующее:
$$
\sup_{x \neq 0} \frac{\|UAVx\|_2}{\|x\|_2} = \sup_{y \neq 0} \frac{\|UAy\|_2}{\|y\|_2} = \sup_{y \neq 0} \frac{\|Ay\|_2}{\|y\|_2} = \|A\|_2
$$
второе равенство выполняется по той же причине. Докажем утверждение для фробениусовой нормы:
$$
\|UAV\|_F^2 = \mathrm{tr}(UAV) = \mathrm{tr}(UU\Lambda V^*V) = \mathrm{tr}(U^*U\Lambda V^*V) = \mathrm{tr}(\Lambda) = \|A\|_F^2
$$   
### Задача 8
Рассмотрим последнюю колонку такой матрицы $A \in \mathbb{R}^{n \times n}$. В этой колонке единственный ненулевой элемент - последний $a_{n, \ n}$. В предпоследней колонке, для выполнения ортогональности с последней колонкой необходимо, что бы элемент $a_{n, \ n-1} = 0$, так что единственный ненулевой элемент - $a_{n - 1, \ n - 1}$. Применяя такую же логику, можно доказать, что для $i$-той колонки единственный ненулевой элемент - $a_{i, \ i}$. Получаем, что такая матрица гарантированно диагональна.
### Задача 9
Для матрицы $S$ справедливо следующее:
$$
\lambda v = Sv
$$
где $\lambda$ и $v$ - соответствующие собственное значение и вектор. Домножим обе части на $v^\top$:
$$
\lambda v^\top v = v^\top Sv
$$
Заметим, что
$$
v^\top Sv = (S^\top v)^\top v = -v^\top Sv
$$
это возможно, только если $\lambda$ имеет нулевую действительную часть. Если матрица $I - S$ необратима, значит существует ненулевой $x$ такой, что $x(I - S) = 0 \iff x = Sx$ - что невозможно, так как все собственные значения $S$ имеют нулевую действительную часть. Тогда $S$ обратима. Проверим ортогональность матрицы $B = (I - S)^{-1}(I + S)$: если $B$ ортогональна, справедливо
$$
BB^\top = I \\ B = B^{-\top} \\ (I - S)^{-1}(I + S) = ((I - S)^{-1}(I + S))^{-\top}
$$
Рассмотрим правую часть выражения:
$$
((I - S)^{-1}(I + S))^{-\top} = ((I + S)^{-1}(I - S))^{\top}
$$
Заметим, что по определению $S$ верно
$$
(I - S)^\top = I^\top - S^\top = I + S
$$
тогда получаем
$$
B^\top = (I - S)^{-1}(I + S) = B
$$

Для таких моделей генерации должна стать ещё быстрее и число параметров также должно ещё сильнее снизиться.



```python
# С помощью этого класса происходит сжатие линейных слоёв
class SVDLinear(nn.Module):
    def __init__(self, orig_layer: nn.Module, rank: int):
        super().__init__()

        if isinstance(orig_layer, nn.Linear):
            u, s, v = torch.linalg.svd(orig_layer.weight.T, full_matrices=False)
            self.bias = nn.Parameter(orig_layer.bias)
            self.rank = rank

            self.u = nn.Parameter(u[:, :rank])
            self.s = nn.Parameter(s[:rank])
            self.v = nn.Parameter(v[:rank])

        # Для большей эффективности, аппроксимации нового меньшего ранга
        # могут быть получены сжатием существующего разложения
        elif isinstance(orig_layer, SVDLinear):
            if rank > orig_layer.rank:
                raise ValueError("Cannot prune more than the original rank.")
            self.u = nn.Parameter(orig_layer.u[:, :rank])
            self.s = nn.Parameter(orig_layer.s[:rank])
            self.v = nn.Parameter(orig_layer.v[:rank])
            self.bias = nn.Parameter(orig_layer.bias)
            self.rank = orig_layer.rank

        else:
            raise NotImplementedError

    def forward(self, x):
        # x @ W.T = x @ (U @ S @ V)
        return ((x @ self.u) * self.s) @ self.v + self.bias
```

```python
# Возвращает размер модели в мегабайтах
def get_model_size(model: nn.Module) -> float:
    param_size = sum(p.nelement() * p.element_size() for p in model.parameters())
    buffer_size = sum(b.nelement() * b.element_size() for b in model.buffers())

    total_size_mb = (param_size + buffer_size) / (1024**2)
    return total_size_mb
```

```python
# Применяет SVD ко всем MLP блокам модели
def prune_mlp_layers(model: nn.Module, rank: int):
    for block in model.transformer.h:
        block.mlp.c_fc = SVDLinear(block.mlp.c_fc, rank)
        block.mlp.c_proj = SVDLinear(block.mlp.c_proj, rank)
```

```python
def timeit(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        return result, end - start

    return wrapper


# Функция для получения генерации модели по промпту
@timeit
@torch.no_grad()
def get_model_output(
    model: nn.Module,
    tokenizer: AutoTokenizer,
    input: str,
    max_new_tokens: int = 100,
    **generation_kwargs,
) -> List[str]:
    generation_kwargs["max_new_tokens"] = max_new_tokens
    input_tokens = tokenizer(input, return_tensors="pt").to(model.device)

    ids_output = model.generate(**input_tokens, **generation_kwargs)
    text_output = [
        tokenizer.decode(ids_output[i], skip_special_tokens=True)
        for i in range(len(ids_output))
    ]
    return text_output
```

```python
# Функция для тестирования модели с записью времени генерации, размера модели
def test_model(
    model: nn.Module,
    tokenizer: AutoTokenizer,
    test_prompts: List[str],
    max_new_tokens: int = 100,
    **generation_kwargs,
) -> Dict[str, Any]:
    model_outputs = []
    total_time = 0

    for prompt in test_prompts:
        model_output, execution_time = get_model_output(
            model, tokenizer, prompt, max_new_tokens=max_new_tokens, **generation_kwargs
        )
        model_outputs.extend(model_output)
        total_time += execution_time

    return dict(
        model_size=get_model_size(model),
        total_time=total_time,
        model_outputs=model_outputs,
    )
```

```python
# Для теста будем брать 4 промпта для генерации функций на python
test_prompts = [
    "Implementation of Insertion Sort in Python:\n\ndef insertion_sort(arr: list[float]) -> list[float]:",
    "Function for matrix multiplication in Python:\n\ndef matrix_multiply(A: list[list[float]], B: list[list[float]]) -> list[list[float]]:",
    "Implementation of Dijkstra's algorithm in Python:\n\nimport heapq\n\ndef dijkstra(graph: dict, start: int, end: int) -> int:",
    "Solution of Knapsack Problem with Dynamic Programming in Python:\n\ndef knapsack_dp(n_items: int, max_weight: int, item_weights: list[int], item_values: list[int]) -> int:",
]

# Такие размерности сжатия будут оцениваться
# "original" - модель без применения сжатия
ranks = ["original", 1600, 1500, 1400, 1300, 1200, 1100]
```

```python
results = []
# Для каждого ранга сжимаем модель и тестируем на промптах,
# записывая результаты в переменную results
for rank in ranks:
    if rank != "original":
        print(f"Pruning model with rank {rank}")
        prune_mlp_layers(model, rank)

    print(f"Testing model with rank {rank}")
    metrics = test_model(
        model,
        tokenizer,
        test_prompts,
        max_new_tokens=500,
        pad_token_id=None,
        # Модель продолжит генерировать после вывода <EOS> токена.
        # Это нужно для точного результата по времени генерации
        eos_token_id=None,
        do_sample=True,
        top_k=5,
        num_beams=30,
        no_repeat_ngram_size=4,
    )
    metrics["rank"] = rank
    results.append(metrics)
```

```python
ranks = [result["rank"] for result in results]
times = [result["total_time"] for result in results]
sizes = [result["model_size"] for result in results]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

ax[0].bar(list(map(str, ranks)), times)
ax[0].set_ylabel("Execution time, s")
ax[0].tick_params(axis="x", rotation=45)
ax[0].set_title("Execution time wrt SVD rank")

ax[1].bar(list(map(str, ranks)), sizes)
ax[1].set_ylabel("Model size, MB")
ax[1].tick_params(axis="x", rotation=45)
ax[1].set_title("Model size wrt SVD rank")

plt.tight_layout()
plt.show()
```

```python
for result in results:
    print("-" * 50)
    print(f"Example generation for rank {result['rank']}:")
    print(result["model_outputs"][0])
    print("-" * 50)
```
